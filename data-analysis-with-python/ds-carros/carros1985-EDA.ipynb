## ANOVA: Análise de Variância

<p><b>Pontuação do teste F</b>: A ANOVA assume que as médias de todos os grupos são iguais, calcula o quanto as médias reais se desviam dessa suposição e relata isso como a pontuação do teste F. Uma pontuação maior significa que há uma diferença maior entre as médias.</p>
<p><b>Valor P</b>: O valor P indica o quão estatisticamente significativo é o valor calculado do nosso teste.</p>
<p>Se nossa variável de preço estiver fortemente correlacionada com a variável que estamos analisando, esperamos que a ANOVA retorne uma pontuação de teste F considerável e um valor P pequeno.</p>

## Tração.

```python
grupo_test2=df_gptest[['tracao', 'preco']].groupby(['tracao'])
grupo_test2.head(2)
```

```python
df_gptest
```

Podemos obter os valores do grupo usando o método "get_group".

```python
grupo_test2.get_group('4wd')['preco']
```

Podemos usar a função 'f_oneway' no módulo 'stats' para obter a <b>pontuação do teste F</b> e o <b>valor P</b>.

```python
# ANOVA
f_val, p_val = stats.f_oneway(grupo_test2.get_group('fwd')['preco'], grupo_test2.get_group('rwd')['preco'], grupo_test2.get_group('4wd')['preco'])  
 
print( "Resultato ANOVA: F=", f_val, ", P =", p_val)   
```


Este é um ótimo resultado, com uma grande pontuação no teste F, indicando uma forte correlação, e um valor P quase igual a 0, o que implica uma significância estatística quase certa. Mas isso significa que todos os três grupos testados têm essa alta correlação?

Vamos examiná-los separadamente.

#### fwd and rwd


```python
f_val, p_val = stats.f_oneway(grupo_test2.get_group('fwd')['preco'], grupo_test2.get_group('rwd')['preco'])  
 
print( "Resultato ANOVA: F=", f_val, ", P =", p_val)  
```

#### 4wd and rwd


```python
f_val, p_val = stats.f_oneway(grupo_test2.get_group('4wd')['preco'], grupo_test2.get_group('rwd')['preco'])  
   
print( "Resultato ANOVA: F=", f_val, ", P =", p_val)  
```

Observamos que a ANOVA para as categorias 4WD e FWD resulta em um valor P elevado > 0,1, portanto, a pontuação do teste F calculada não é estatisticamente significativa. Isso sugere que não podemos rejeitar a suposição de que as médias desses dois grupos são iguais, ou seja, não podemos concluir que a diferença na correlação seja significativa.
