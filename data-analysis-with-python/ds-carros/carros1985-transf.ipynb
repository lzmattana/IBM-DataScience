```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
```

```python
# adicionando identificadores
cabecalhos = ["classificacao", "perdas-normalizadas", "fabricante", "tipo-de-combustivel", "aspiracao", "numero-de-portas", "estilo-do-carro",
              "tipo-de-tracao", "localizacao-do-motor", "base-das-rodas", "comprimento", "largura", "altura", "peso", "tipo-de-motor",
              "numero-de-cilindros", "tamanho-do-motor", "sistema-de-combustivel", "diametro-do-cilindro", "curso-do-pistao",
              "taxa-de-compressao", "potencia", "rpm-maximo", "mpg-na-cidade", "mpg-na-estrada", "preco"]
```

<h4>Calcular a mÃ©dia para perdas-normalizadas</h4>


```python
avg_norm_loss = df["perdas-normalizadas"].astype("float").mean(axis=0)
print("A mÃ©dia das perdas-normalizadas: ", avg_norm_loss)
```

```python
df["perdas-normalizadas"].replace(np.nan, avg_norm_loss, inplace=True)
```

<h4>Calcular a mÃ©dia para diametro-do-cilindro.</h4>

```python
avg_bore=df['diametro-do-cilindro'].astype('float').mean(axis=0)
print("A mÃ©dia do diametro-do-cilindro: ", avg_bore)
```

```python
df["diametro-do-cilindro"].replace(np.nan, avg_bore, inplace=True)
```

<h4>Calcular a mÃ©dia para curso do pistÃ£o</h4>

```python
avg_stroke=df['curso-do-pistao'].astype('float').mean(axis=0)
print("A mÃ©dia do curso do pistÃ£o: ", avg_stroke)
```

```python
df['curso-do-pistao'].replace(np.nan, avg_stroke, inplace=True)
```

<h4>Calcular a mÃ©dia para cavalo de potencia.</h4>

## PadronizaÃ§Ã£o de dados

<p>Nota-se que o consumo estÃ¡ em milhas por galÃ£o, usaremos o padrÃ£o internacional Km/L.</p>
A formula de conversÃ£o: km/l = 235 / mpg


```python
df['mpg-na-cidade'] = 235/df['mpg-na-cidade']
df['mpg-na-estrada'] = 235/df['mpg-na-estrada']
df.rename(columns={'mpg-na-cidade': 'kml-cidade'}, inplace=True)
df.rename(columns={'mpg-na-estrada': 'kml-estrada'}, inplace=True)
df.head()
```

## NormalizaÃ§Ã£o de dados

Normalizaremos comprimento, largura e altura.

```python
df['largura'] = df['largura']/df['largura'].max()
df['comprimento'] = df['comprimento']/df['comprimento'].max()
df['altura'] = df['altura']/df['altura'].max() 
df[['comprimento', 'largura', 'altura']].head()
```

<h2>Agrupamento (Binning)</h2>

<p>No conjunto de dados, "potÃªncia" Ã© uma variÃ¡vel de valor real que varia de 48 a 288 e possui 59 
valores Ãºnicos. Vamos utilizar os valores de agrupamento alta potÃªncia, 
potÃªncia mÃ©dia e baixa potÃªncia (3 tipos)</p>

```python
df["potencia"]=df["potencia"].astype(int, copy=True)
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.hist(df["potencia"])

plt.pyplot.xlabel("potencia")
plt.pyplot.ylabel("valor")
plt.pyplot.title("potencia e valor")
```

```python
bins = np.linspace(min(df["potencia"]), max(df["potencia"]), 4)
bins
```

```python
nomes_grupos = ['baixo', 'medio', 'alto']
```

```python
df['potencia-binned'] = pd.cut(df['potencia'], bins, labels=nomes_grupos, include_lowest=True )
df[['potencia','potencia-binned']].head(10)
```

```python
df["potencia-binned"].value_counts()
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
pyplot.bar(nomes_grupos, df["potencia-binned"].value_counts())

plt.pyplot.xlabel("potencia")
plt.pyplot.ylabel("valor")
plt.pyplot.title("potencia e valor")
```

<h2>VariÃ¡vel Indicadora (ou VariÃ¡vel Dummy)</h2>
<p>
    A coluna "tipo-de-combustÃ­vel" possui dois valores Ãºnicos: "gasolina" ou "diesel". A regressÃ£o nÃ£o compreende palavras, apenas nÃºmeros. Para usar esse atributo em anÃ¡lises de regressÃ£o, vocÃª pode converter "tipo-de-combustÃ­vel" em variÃ¡veis indicadoras.
</p>
<p>
'get_dummies' do Pandas para atribuir valores numÃ©ricos a diferentes categorias de tipo de combustÃ­vel.
</p>

```python
df.columns
```

```python
df["tipo-de-combustivel"].unique()
```

```python
dummy_variable_1 = pd.get_dummies(df["tipo-de-combustivel"])
dummy_variable_1.head()
```

```python
dummy_variable_1.rename(columns={'gas':'combustivel-gas', 'diesel':'combustivel-diesel'}, inplace=True)
dummy_variable_1.head()
```

```python
df = pd.concat([df, dummy_variable_1], axis=1)

df.drop("tipo-de-combustivel", axis = 1, inplace=True)
```

```python
df.head()
```

```python
dummy_variable_2 = pd.get_dummies(df['aspiracao'])

dummy_variable_2.rename(columns={'std':'aspiracao-normal', 'turbo': 'aspiracao-turbo'}, inplace=True)

dummy_variable_2.head()
```

```python
df = pd.concat([df, dummy_variable_2], axis=1)

df.drop('aspiracao', axis = 1, inplace=True)
```
