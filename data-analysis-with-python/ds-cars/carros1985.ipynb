```python
import pandas as pd
import numpy as np
```

```python
# Prévia análise mostra que dataset vem seu cabeçalho
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"
df = pd.read_csv(url, header=None)
df.head(5)
```

```python
df.tail(5)
```

```python
# adicionando identificadores
cabecalhos = ["classificacao", "perdas-normalizadas", "fabricante", "tipo-de-combustivel", "aspiracao", "numero-de-portas", "estilo-do-carro",
              "tipo-de-tracao", "localizacao-do-motor", "base-das-rodas", "comprimento", "largura", "altura", "peso", "tipo-de-motor",
              "numero-de-cilindros", "tamanho-do-motor", "sistema-de-combustivel", "diametro-do-cilindro", "curso-do-pistao",
              "taxa-de-compressao", "potencia", "rpm-maximo", "mpg-na-cidade", "mpg-na-estrada", "preco"]
df.columns = cabecalhos
df.head(5)
```

<p>
Precisaremos subistituir '?' por NaN e dropna() para valores faltantes.
</p>

```python
df1=df.replace('?',np.NaN)
df=df1.dropna(subset=["preco"], axis=0)
df.head(10)
```
